# Exploration interactive des données PhishTank

Ce notebook explore le flux public PhishTank enregistré dans `../data/urls_phishtank_brutes.csv`.
PhishTank fournit un flux public de phishing vérifié, donc ce notebook et le script associé fonctionnent sans clé API.

```python
from pathlib import Path
from io import StringIO
from urllib.request import Request, urlopen

import pandas as pd

output_path = Path("../data/urls_phishtank_brutes.csv")
phishtank_url = "https://data.phishtank.com/data/online-valid.csv"
request = Request(phishtank_url, headers={"User-Agent": "phishtank-notebook/1.0"})

if output_path.exists():
    df = pd.read_csv(output_path)
else:
    with urlopen(request) as response:
        csv_text = response.read().decode("utf-8", errors="replace")
    df = pd.read_csv(StringIO(csv_text))
    output_path.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(output_path, index=False)

display(df.head())
```

# Exploration interactive des données

Ce notebook répond à la question : à quoi ressemblent mes données ?

Je vais charger le CSV sauvegardé, inspecter sa forme, ses types, ses statistiques et les valeurs manquantes.

```python
import pandas as pd

df = pd.read_csv("../data/urls_brutes.csv", header=None, names=[
    "id",
    "dateadded",
    "url",
    "url_status",
    "last_online",
    "threat",
    "tags",
    "urlhaus_link",
    "reporter",
])
df.head()
```

```python
df["tags"].value_counts().head(10)
```

```python
df["tags"].str.contains("phishing", case=False, na=False).sum()
```

## Observations rapides

- Le fichier contient 21 325 URLs au total.
- Une URL est taguée `phishing`.
- Les valeurs manquantes se concentrent surtout dans `last_online` (4 386 valeurs manquantes) et `tags` (1 300 valeurs manquantes).
- Les tags les plus fréquents sont dominés par des familles comme `LuaJIT-loader,SmartLoader,SmartLoader-MaaS`, `32-bit,elf,mips,Mozi` et `elf,mirai,ua-wget`.
- Le CSV brut a été sauvegardé sans en-têtes, donc les colonnes ont dû être renommées pour rendre l’exploration interprétable.

## Décision : changement de source

Le flux URLhaus s'est révélé peu adapté au phishing : sur 21 325 URLs, une seule est taguée `phishing`, alors que les tags dominants correspondent surtout à des familles de malware et de loaders.
Le projet bascule donc vers PhishTank, une source spécialisée dans le phishing vérifié par la communauté.

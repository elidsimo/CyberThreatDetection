## Observations sur le dataset AbuseIPDB

- 1000 adresses IP recuperees (limite du plan gratuit par requete).
- Toutes ont un score de confiance de 100 (abuseConfidenceScore), 
  min = max = 100, ecart-type = 0.
- Le parametre confidenceMinimum=90 utilise dans le script est en realite 
  redondant : l'API AbuseIPDB trie par score decroissant, donc les 1000 
  premieres entrees de la blacklist sont deja toutes a confiance maximale.
- Cette blacklist represente donc uniquement les IPs les "pires" 
  (confiance absolue), pas un echantillon representatif de toutes 
  les IPs suspectes.
